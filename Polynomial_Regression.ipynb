# Importing Some Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
```

# Make a Polynomial Dataset

```python
noise = np.random.rand(100,1)
x = 8*np.random.rand(100,1)-4
y = 0.6*(x**2) + 0.4*x - 2*noise
```

```python
plt.scatter(x,y);
```

# Predicting using Linear Regression

```python
clf = LinearRegression()
clf.fit(x,y)
y_pred = clf.predict(x)
plt.plot(x,y_pred,label="predicted")
plt.scatter(x,y,label="Actual")
plt.legend()
plt.show()
```

```python
y_pred = clf.predict(x)
print(r2_score(y,y_pred))
print(mean_squared_error(y,y_pred))
```

# Transform the feature in higher degree

```python
poly = PolynomialFeatures(degree=2)
x_transform = poly.fit_transform(x)
x_transform.shape
```

```python
clf.fit(x_transform,y)
y_pred1 = clf.predict(x_transform)
plt.scatter(x,y_pred1,label="Predicted after poly_feature")
plt.scatter(x,y,label = "Actual")
plt.legend()
plt.show()
```

```python
y_pred = clf.predict(x_transform)
print(r2_score(y,y_pred))
print(mean_squared_error(y,y_pred))
```

```python
# Befor using polynomial feature r2_score is 0.07 but after transform 
# the feature it exceeds to 0.96
```
